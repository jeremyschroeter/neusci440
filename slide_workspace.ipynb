```python
import matplotlib.pyplot as plt
import numpy as np
import neuscitk as ntk
```

```python
dataset = ntk.LabChartDataset('assets/datasets/ntk_tut_data.mat')
```

```python
block = dataset.get_block(0)
```

```python
import numpy as np

def generate_spike_trains(duration, num_neurons, firing_rate):
    spike_trains = []
    for _ in range(num_neurons):
        spike_train = np.random.rand(duration) < firing_rate
        spike_trains.append(spike_train)
    return spike_trains

spikes = generate_spike_trains(1000, 1, 0.03)[0]
sinwave = np.sin(np.linspace(0, 20 * 2*np.pi, 1000)) * 20
```

```python
plt.rcParams['figure.dpi'] = 300
fig, ax = plt.subplots(2)

ax[0].plot(spikes, color='w')
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].spines['bottom'].set_visible(False)
ax[0].spines['left'].set_visible(False)
ax[1].plot(sinwave, color='w')
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['bottom'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].set_ylim(-25, 25)

fig.tight_layout()

plt.savefig('assets/figures/spike_train.png', transparent=True)
```
